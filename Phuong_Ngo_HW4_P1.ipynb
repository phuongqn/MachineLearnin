```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
cc=pd.read_csv('/Users/phuongqn/Desktop/INF552/Homework/Homework 4 Data/communities.data', header=None)
cc
```

```python
ccnames=pd.read_excel('/Users/phuongqn/Desktop/INF552/Homework/Homework 4 Data/ccnames.xls', columms='names')
ccnames=ccnames.names.str.split(' ', 3, expand=True)
ccnames=ccnames.drop(columns=[0,2])
ccnames=ccnames.values.tolist()

cc_names=[]
for i in ccnames:
    for j in i:
        cc_names.append(j)
print(cc_names)
```

```python
cc.columns= cc_names
cc=cc.drop(columns=['state','county','community','communityname','fold'])
```

```python
for column in cc:
    cc[column] = pd.to_numeric(cc[column],errors='coerce')
cc = cc.replace('?',np.NaN)
cc
```

```python
cc_copy=cc.copy()

train_cc=cc_copy[:1495]
test_cc=cc_copy[1495:]
test_cc
```

```python
from sklearn.impute import SimpleImputer
mean_imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

mean_imputer1 = mean_imputer.fit(test_cc)
imputed_t = mean_imputer1.transform(test_cc.values)        
imputed_test= pd.DataFrame(imputed_t, columns=test_cc.columns)  

mean_imputer2 = mean_imputer.fit(train_cc)
imputed_tr=  mean_imputer2.transform(train_cc.values)
imputed_train= pd.DataFrame(imputed_tr, columns=train_cc.columns)
```

```python
imputed_train
```

```python
y_train=imputed_train['ViolentCrimesPerPop'].values
y_test=imputed_test['ViolentCrimesPerPop'].values

X_train=imputed_train.drop(columns = ['ViolentCrimesPerPop'])
X_test=imputed_test.drop(columns = ['ViolentCrimesPerPop'])

X_train

imputed_X=pd.concat([X_train,X_test], ignore_index=True)
imputed_X
```

```python
corr = imputed_X.corr()

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

f, ax = plt.subplots(figsize=(25,25))
cmap = sns.diverging_palette(220, 10, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, cbar_kws={"shrink": .5})
```

```python
print(corr)
```

```python
imputed_y=np.append(y_test,y_train)
imputed_cc=imputed_X.copy()
imputed_cc['ViolentCrimesPerPop'] = imputed_y
imputed_cc
```

```python
CV = imputed_cc.std()/imputed_cc.mean()
CVcc = pd.DataFrame({'Attribute':CV.index, 'CV':CV.values})
CVcc
```

```python
import math
sorted_cv = CVcc.sort_values(by=['CV'],ascending=False)
sorted_cv
n = math.ceil(math.sqrt(128))
top_n = sorted_cv.head(n)
print(top_n)
```

```python
top_n=top_n.set_index('Attribute',drop=True)
top_n_names=list(top_n.index.values)
top_n_names
```

```python
import statsmodels.api as sm
regression = sm.OLS(y_train, X_train).fit()
print(regression.summary())
#fit linear model
```

```python
test_pred = regression.predict(X_test)
test_er = sum((y_test-test_pred)**2)/(len(test_pred))
train_pred = regression.predict(X_train)
train_er = sum((y_train-train_pred)**2)/(len(train_pred))
print('Train MSE : ',train_er)
print('Test MSE : ',test_er)
```

```python
from sklearn.linear_model import RidgeCV
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import Ridge
from sklearn.linear_model import LassoCV
from sklearn.linear_model import Lasso
a_ridge = [1e15, 1e10, 1e8, 1e4, 1e3,1e2, 5, 10, 20, 0.1, 1, 10, 100]
ridge_reg_cv = RidgeCV(alphas=a_ridge, cv = KFold(n_splits=10, shuffle = True)).fit(X_train, y_train)
print('Best lambda chosen by cross validation: ',ridge_reg_cv.alpha_)
```

```python
rr = Ridge(alpha=1.0)
rr = rr.fit(X_train,y_train)
ridge_tr_pred = rr.predict(X_train)
ridge_t_pred = rr.predict(X_test)
ridge_tr_error = sum((y_train-ridge_tr_pred)**2)/(len(ridge_tr_pred))
ridge_t_error = sum((y_test-ridge_t_pred)**2)/(len(ridge_t_pred))
print('Training error on ridge regression : ',ridge_tr_error)
print('Test error on ridge regression : ',ridge_t_error)
```
